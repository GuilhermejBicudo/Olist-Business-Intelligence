## Dataset Summary

```python
overview = []

for name, df in datasets.items():
    overview.append(
        {
            "Dataset": name,
            "Rows": df.shape[0],
            "Columns": df.shape[1],
            "Missing Values": df.isnull().sum().sum(),
            "Duplicate Rows": df.duplicated().sum(),
        })
    
overview = pd.DataFrame(overview)

overview
```

## Orders Dataset Exploration

In this section, we explore the structure of the `orders` table, identifying its columns, data types, and general characteristics before performing any transformations.

```python
orders = datasets["orders"]
```

```python
orders.head()
```

```python
orders["order_status"].unique()
```

```python
orders["order_status"].nunique()
```

```python
orders["order_status"].value_counts()
```

### Key Findings

- The `order_status` column contains **8 unique categories**, indicating a low-cardinality categorical variable.
- Most orders are marked as `delivered`, suggesting a high order fulfillment rate.
- A small proportion of orders were canceled or became unavailable.
- Only a few orders remain in the initial stages (`created` and `approved`), which is expected because this dataset represents historical data.

### Candidate Primary Key Validation

```python
rows = orders.shape[0]
unique_ids = orders["order_id"].nunique()
null_ids = orders["order_id"].isna().sum()

print(f"Rows: {rows}")
print(f"Unique IDs: {unique_ids}")
print(f"Null IDs: {null_ids}")
```

#### Key Findings

- `order_id` contains no missing values.
- The number of unique `order_id` values matches the total number of rows.
- Therefore, `order_id` is a strong candidate for the primary key of the `orders` table.
